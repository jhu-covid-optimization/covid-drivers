```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
analysis_rucc_1 = pd.read_csv('../covid19_epi_model/data/05-analysis_df_rucc_1.csv')
analysis_rucc_2 = pd.read_csv('../covid19_epi_model/data/05-analysis_df_rucc_2.csv')
analysis_rucc_3 = pd.read_csv('../covid19_epi_model/data/05-analysis_df_rucc_3.csv')
```

```python
analysis_rucc_1['rucc'] = np.ones(len(analysis_rucc_1))
analysis_rucc_2['rucc'] = np.ones(len(analysis_rucc_2)) * 2
analysis_rucc_3['rucc'] = np.ones(len(analysis_rucc_3)) * 3
```

```python
analysis_df = pd.concat((analysis_rucc_1, analysis_rucc_2, analysis_rucc_3), axis=0)
```

```python
sns.set(font_scale=2, context='paper', rc={'legend.title_fontsize': 20})
vars = ['PC1', 'PC2', 'PC3', 'PC4', 'beds_log', 'hospct_log']
hue = 'rucc'
```

```python
g = sns.pairplot(
    analysis_df[vars + [hue]].drop_duplicates(),
    vars=vars,
    hue=hue,
)
```

```python
g.savefig("../figures/epi_model_analyses_pairplot.pdf")
```
